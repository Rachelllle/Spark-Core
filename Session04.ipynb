```python
from pyspark import SparkConf, SparkContext

# Configure Spark. Use getOrCreate to handle multiple executions.
conf = SparkConf().setAppName("MonApp").setMaster("local")
sc = SparkContext.getOrCreate(conf=conf)

# Création du RDD depuis la liste de tuples
data = [
    ("ERROR", 1),
    ("INFO", 1),
    ("ERROR", 1),
    ("WARNING", 1),
    ("INFO", 1),
    ("ERROR", 1)
]

rdd = sc.parallelize(data)
```

```python

# Count occurrences per level
rdd_count = rdd.reduceByKey(lambda a, b: a + b)
# → [("ERROR", 3), ("INFO", 2), ("WARNING", 1)]

# Find the most frequent level
most_frequent = rdd_count.reduce(lambda a, b: a if a[1] > b[1] else b)

print("Occurrences per level:", rdd_count.collect())
print("Most frequent level:", most_frequent)
```
