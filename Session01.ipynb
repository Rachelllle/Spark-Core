```python
from pyspark.sql import SparkSession

# 1️⃣ Créer une session Spark
spark = SparkSession.builder \
    .appName("Intro-Spark") \
    .master("local[*]") \
    .getOrCreate()

print("Version Spark :", spark.version)

# 2️⃣ Créer un RDD avec les nombres de 1 à 10
sc = spark.sparkContext
rdd = sc.parallelize(range(1, 11))
print("Contenu du RDD :", rdd.collect())

# 3️⃣ Calculer la somme des nombres
print("Somme :", rdd.sum())

# 4️⃣ Créer un RDD avec les carrés
rdd_carre = rdd.map(lambda x: x**2)
print("Carrés :", rdd_carre.collect())

# 5️⃣ Calculer la somme des carrés
print("Somme des carrés :", rdd_carre.sum())

# 6️⃣ Arrêter Spark
spark.stop()
```
